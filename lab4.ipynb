В качестве примера использования классов, __repr__ представлена программа для топологической декомпозиции графа. На вход программа принимает файл с матрицей смежности графа, на выход выдает топологическую декомпозицию графа.

```python
from array import *
import pandas as pd
import numpy as np
path = "./smmatr3"

class Edge:
    v1 = 0
    v2 = 0
    def __init__(self, v1, v2):
        self.v1 = v1
        self.v2 = v2
    def __repr__(self) -> str:
        return f'[{self.v1+1}, {self.v2+1}]'

class Subgraph:
    V = list[int]()
    def __init__(self):
        self.V = list(0)
    def __init__(self,  V: list[int]):
        self.V = V
    def __repr__(self) -> str:
        return str(list(map(lambda x: x+1, self.V)))


def topologicalDecomposition( countV: int, E: list[Edge], Sub: list[Subgraph]):
    notUsedV = list[int]() #список еще не использованных вершин
    for i in range(countV):
        notUsedV.append(i)
    for i in range(countV):
        while (len(notUsedV) > 0):
            R = list[int]() #достижимое множество
            R.append(notUsedV[0])
            Q = list[int]() #контрдостижимое множество
            Q.append(notUsedV[0])
            color = np.zeros(vCount,int) 
            for i in range(1, len(notUsedV)):
                for k in range(countV):
                    try: 
                        notUsedV.index(k)
                        color[k] = 1
                    except:
                        color[k] = 2
                if (DFS(notUsedV[0], notUsedV[i], E, color)): R.append(notUsedV[i])
                for k in range(countV):
                    try: 
                        notUsedV.index(k)
                        color[k] = 1
                    except:
                        color[k] = 2
                if (DFS(notUsedV[i], notUsedV[0], E, color)): Q.append(notUsedV[i])
            intersection = list(set(R) & set(Q))
            print(f'Step {i}')
            print("R= "+ str(list(map(lambda x: x+1, R))))
            print("Q= " + str(list(map(lambda x: x+1, Q))))
            print("V= "+ str(list(map(lambda x: x+1, intersection))))
            Sub.append(Subgraph(intersection))
            for i in range(len(intersection)):
                notUsedV.remove(intersection[i])
            print("G= "+ str(list(map(lambda x: x+1, notUsedV))))

def DFS(u: int, endV: int, E: list[Edge], color: array) -> bool:
    color[u] = 2
    if (u == endV):
        return True
    for w in range(len(E)):
        if (color[E[w].v2] == 1 and E[w].v1 == u):
            if (DFS(E[w].v2, endV, E, color)): return True
            color[E[w].v2] = 1
    return False
        
def findEdges(data) -> list[Edge]:
    edges = list[Edge]()
    for i in range(len(data)):
        for j in range(len(data[i])):
            if (int(data.iloc[j, i]) == 1):
                edges.append(Edge(v1=j, v2=i))
    return edges


if __name__ == "__main__":
    data = pd.read_csv(path, sep=", ", header=None, engine='python')
    print("input matrix: ")
    print (data)
    data[len(data.columns)-1] = data[len(data.columns)-1].str.replace(',', '')
    print ("modifed data: ")
    print(data)

    edges = list[Edge]()
    edges = findEdges(data)
    print("edges: ")
    print(edges)
    vCount = len(data.columns)
    Subgr = list[Subgraph]()
    topologicalDecomposition(vCount, edges, Subgr)
    print("Subgraphs: ")
    print(Subgr)
```
